```python
import matplotlib.pyplot as plt
import networkx as nx

def read_input():
    with open("input.txt") as f:
        return sorted([i.split('-') for i in f.read().splitlines()], reverse=True)

cons = read_input()
graph = nx.from_edgelist(cons)
nx.draw(
    graph,
    with_labels=False,
    node_size=10,
    node_color='blue',
    edge_color='gray',
    width=0.5
)
plt.show()
```

```python
# Find all cliques of size 3 (triangles) in the graph
triangles = [clique for clique in nx.enumerate_all_cliques(graph) if len(clique) == 3]
print(len(triangles))
```

```python
total = 0
for cl in triangles:
    ok = False
    for cmp in cl:
        if cmp.startswith('t'):
            ok = True
            print(cl)
            break
    total += int(ok)
print(total)
```

```python
max(nx.find_cliques(graph), key=len)
```

```python
largest = max([clique for clique in nx.enumerate_all_cliques(graph)], key=len)
```

```python
node_colors = ['red' if node in largest else 'blue' for node in graph.nodes()]

nx.draw(
    graph,
    with_labels=False,
    node_size=10,
    node_color=node_colors,
    edge_color='gray',
    width=0.5
)
plt.show()
```
